# Toy Dataset 2 predicting probabilities

```python
import torch as th
```

```python
input_tensor = th.tensor([[5.8],[6.0],[5.5],[4.5],[4.1],[3.5]],requires_grad=True)
target_tensor = th.tensor([[1.0],[1],[1],[0],[0],[0]],requires_grad=False)
```

```python
from torch.optim import optimizer
from torch import optim
from torch import nn
import torch.nn.functional as F
import numpy
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(1,1)
    def forward(self,x):
        x = th.sigmoid(self.fc(x))
        return x
```

```python
model = Classifier()
criterion = nn.BCEWithLogitsLoss()

optimizer = optim.SGD(model.parameters(), lr =0.4,momentum=0.8)

for i in range(10000):
    optimizer.zero_grad()
    predict_tensor = model(input_tensor)
    predict_tensor = th.log(predict_tensor)
    loss = criterion(predict_tensor,target_tensor)
    loss.backward()
    optimizer.step()
    print(loss.data)
```
