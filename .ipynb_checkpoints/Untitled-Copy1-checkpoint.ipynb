```python
import torch as th
```

```python
input_tensor = th.tensor([[4.2],[5.6],[4.7],[1.2],[1.7],[3.5]],requires_grad=True)
target_tensor = th.tensor([[1.0],[1],[1],[0],[0],[0]],requires_grad=True)
```

```python
from torch.optim import optimizer
from torch import optim
from torch import nn
import torch.nn.functional as F
model = nn.Linear(1,1)
```

```python
input_tensor.shape
```

```python
target_tensor.shape
```

```python
target_tensor
```

```python
# target_tensor.squeeze()
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(model.parameters(), lr =0.1)

for i in range(1000):
    optimizer.zero_grad()
    predict_tensor = model(input_tensor)
    # print(predict_tensor)
    loss = criterion(predict_tensor,target_tensor)
    loss.backward()
    optimizer.step()
    print(loss.data)
```

```python
predict = model(th.tensor([0.5]))
```
