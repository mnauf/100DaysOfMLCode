```python
import torch as th
```

```python
input_tensor = th.tensor([[4.2],[5.6],[4.7],[1.2],[1.7],[3.5]],requires_grad=True)
target_tensor = th.tensor([[1],[1],[1],[0],[0],[0]])
```

```python
from torch.optim import optimizer
from torch import optim
from torch import nn
import torch.nn.functional as F
model = nn.Linear(1,2)
```

```python
input_tensor.shape
```

```python
target_tensor.shape
```

```python
target_tensor
```

```python
target_tensor.squeeze()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr =0.1)

for i in range(1000):
    optimizer.zero_grad()
    predict_tensor = model(input_tensor)
    predict_tensor = predict_tensor.view(6,2)
    target_tensor = target_tensor.long()
    loss = criterion(predict_tensor,target_tensor.squeeze())
    loss.backward()
    optimizer.step()
    print(loss.data)
```

```python
predict = model(th.tensor([5.5]))
```
