# Toy Dataset 2 predicting logits

```python
import torch as th
import matplotlib.pyplot as plt
import numpy as np
```

```python
input_tensor = th.tensor([[5.8],[6.0],[5.5],[4.5],[4.1],[3.5]],requires_grad=True)
target_tensor = th.tensor([[1.0],[1],  [1],  [0],  [0],  [0]],requires_grad=False)
input_array = np.array([[5.8],[6.0],[5.5],[4.5],[4.1],[3.5]])
target_array = np.array([[1.0],[1],  [1],  [0],  [0],  [0]])
```

```python
plt.scatter([input_array[i] for i in range(3)],[[0],[0],[0]], s = 50, color = 'green', edgecolor = 'k')
plt.scatter([input_array[i] for i in range(3,6)],[[0],[0],[0]], s = 50, color = 'red', edgecolor = 'k')
```

```python
def display(m, b, color='g--'):
    plt.xlim(3.3,6.3)
    plt.ylim(-0.2,1.2)
    x = np.arange(-10, 10, 0.1)
#     x = th.from_numpy(x)
    plt.plot(x, m*x+b, color)
```

```python
from torch.optim import optimizer
from torch import optim
from torch import nn
import torch.nn.functional as F
model = nn.Linear(1,1)
```

```python
criterion = nn.BCEWithLogitsLoss()

optimizer = optim.SGD(model.parameters(), lr =0.4,momentum=0.8)

for i in range(10000):
    optimizer.zero_grad()
    predict_tensor = model(input_tensor)
    loss = criterion(predict_tensor,target_tensor)
    loss.backward()
    optimizer.step()
    print(loss.data)
    weight = (model.weight.data)[0][0].tolist()
    bias = (model.bias.data)[0].tolist()
#     print(weight)
#     print(bias)
    if i == 0:
        display(weight,bias)
    if i%500 == 0:
        display(weight,bias)
plt.title("Solution boundary")
display(weight,bias,'black')
plt.scatter([input_array[i] for i in range(3)],[[0],[0],[0]], s = 50, color = 'green', edgecolor = 'k')
plt.scatter([input_array[i] for i in range(3,6)],[[0],[0],[0]], s = 50, color = 'red', edgecolor = 'k')
plt.show()
```

# Toy Dataset 2 predicting probabilities

```python
import torch as th
```

```python
input_tensor = th.tensor([[5.8],[6.0],[5.5],[4.5],[4.1],[3.5]],requires_grad=True)
target_tensor = th.tensor([[1.0],[1],[1],[0],[0],[0]],requires_grad=False)
```

```python
from torch.optim import optimizer
from torch import optim
from torch import nn
import torch.nn.functional as F
import numpy
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(1,1)
    def forward(self,x):
        x = th.sigmoid(self.fc(x))
        return x
```

```python
model = Classifier()
criterion = nn.BCEWithLogitsLoss()

optimizer = optim.SGD(model.parameters(), lr =0.4,momentum=0.8)

for i in range(10000):
    optimizer.zero_grad()
    predict_tensor = model(input_tensor)
    predict_tensor = th.log(predict_tensor)
    loss = criterion(predict_tensor,target_tensor)
    loss.backward()
    optimizer.step()
    print(loss.data)
```
